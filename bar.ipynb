```python
import matplotlib.pyplot as plt
```

```python
labels=['A','B','C']
values=[1,4,2]
plt.title("my bar graph")
plt.xlabel("calls")
plt.ylabel("departments")
bars=plt.bar(labels,values,color="red")
bars[0].set_hatch('+')
bars[1].set_hatch('o')
bars[2].set_hatch('*')
plt.figure(figsize=(6,4))

plt.show()
```
